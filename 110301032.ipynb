```python
from PIL import Image, ImageDraw, ImageFont   #讀入pil模組(在圖片上加入文字需要)
import textwrap
```

```python
img = Image.open('123.png') # 開啟圖片
draw = ImageDraw.Draw(img) #將圖片設定為等一下新增文字的目標
#font = ImageFont.truetype('NotoSansTC-Bold.otf', 65) # 給定字型及大小
font_location='/usr/share/fonts/truetype/NotoSansTC-Bold.otf'
font = ImageFont.truetype(font_location, 65)
# '/usr/share/fonts/truetype/NotoSansTC-Bold.otf'
text = input('請輸入你支離滅裂な思考発言!')
para = textwrap.wrap(text, width=8) # 設定每行的文字長度
MAX_W, MAX_H = 1200, 200 #設定文字在圖片上的位置參數
current_h, pad = 145, 20
for line in para: 
    w, h = draw.textsize(line, font=font) 
    draw.text(((MAX_W - w)/1.3, current_h), line, font=font,fill=input("請用rgb(r,g,b)的形式輸入想要的文字顏色，例如rgb(232,43,123)。")) 
    current_h += h + pad 
img.save('1234.png') # 儲存圖片
img2 = Image.open('1234.png') # 顯示圖片
img2
```

```python
from PIL import Image
import cv2
import numpy
```

```python
img = cv2.imread('123.png')  # 影象從OpenCV格式轉換成PIL格式
img2= Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
font = ImageFont.truetype('NotoSansTC-Bold.otf', 65)

text = str.encode('utf8')   # 需要先把輸出的中文字元轉換成Unicode編碼形式
text = input('請輸入你支離滅裂な思考発言!')

draw = ImageDraw.Draw(img2)
para = textwrap.wrap(text, width=8) # 設定每行的文字長度
MAX_W, MAX_H = 1200, 200 #設定文字在圖片上的位置參數
current_h, pad = 145, 20
for line in para: 
    w, h = draw.textsize(line, font=font) 
    draw.text(((MAX_W - w)/1.3, current_h), line, font=font,fill=(0,0,0)) 
    current_h += h + pad

# 轉換回OpenCV格式
img= cv2.cvtColor(numpy.asarray(img2), cv2.COLOR_RGB2BGR) # 轉換回OpenCV格式

cv2_imshow(img)  #  顯示圖片
cv2.waitKey()
```

```python
# 讀取圖檔
img=cv2.imread('123.png')

# 文字內容
text =input("請輸入你支離滅裂な思考発言! ")

# 圖片加入文字方塊
img = ps.putBText(
        img,                            # 原始影像
        text,                            # 文字內容
        text_offset_x = 675,                    # X 軸座標
        text_offset_y = 170,                    # Y 軸座標
        vspace = 10,                       # 縱向空間
        hspace = 10,                       # 橫向空間
        font_scale = 2.0,                     # 字型大小
        background_RGB = (255, 255, 255),           # 背景顏色
        text_RGB = (1, 1, 1)                 # 文字顏色 
        )


cv2_imshow(img) #  顯示圖片
cv2.waitKey(0)
```
